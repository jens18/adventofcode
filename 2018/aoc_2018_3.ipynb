```python
filename = "aoc_2018_3.txt"

# part 1:
def read_file_claims():
    "large reference dataset"
    claims = []
    file = open(filename, "r")
    for line in file:
        raw_claim, dummy, raw_coord, raw_dim = line.split()
        x_start, y_start = raw_coord.strip(':').split(',')
        dim_x, dim_y = raw_dim.split('x')
        claim = raw_claim.strip('#')
    
        # calc abs. coordinates based on start coordinates (x,y) and dimension (x * y)
        x_range = [ int(x_start), int(x_start) + (int(dim_x) -1)]
        y_range = [ int(y_start), int(y_start) + (int(dim_y) -1)]
        
        claims.append([x_range, y_range])
    return claims

def read_test_claims():
    "small sample dataset"
    claims = []
    claims.append([[1,4],[3,6]])
    claims.append([[3,6],[1,4]])
    claims.append([[5,6],[5,6]])
    return claims
    
claims = read_file_claims()
#claims = read_test_claims()
claim_num = len(claims)
    
max_x = 0
max_y = 0

def print_claims(c):
    "print all claims in dict"
    for i in c:
        print(i)
        
def square_inch(x_l, x_r, y_l, y_r):
    return ((x_r - x_l) + 1) * ((y_r - y_l) + 1)
        
def compare_dim(a_l, a_r, b_l, b_r):
    o_l = -1
    o_r = -1

    if a_r >= b_l and b_r >= a_r:
        o_l = b_l
        o_r = a_r
    if a_l <= b_r and a_l >= b_l: 
        o_l = a_l
        o_r = b_r
    if a_l <= b_l and a_r >= b_r: 
        o_l = b_l
        o_r = b_r
    if a_l >= b_l and a_r <= b_r: 
        o_l = a_l
        o_r = a_r
        
    return o_l, o_r      

        
def compare_claims(a, b):
    "compare indiv. a and b claims, return True if overlap exists"
    
    global max_x, max_y
    
    a_x_l = a[0][0]
    a_x_r = a[0][1]
    a_y_l = a[1][0]
    a_y_r = a[1][1]
    
    b_x_l = b[0][0]
    b_x_r = b[0][1]
    b_y_l = b[1][0]
    b_y_r = b[1][1]
    
    # update max_x and max_y
    if a_x_r > max_x:
        max_x = a_x_r
    if a_y_r > max_y:
        max_y = a_y_r    
    
    o_x_l, o_x_r = compare_dim(a_x_l, a_x_r, b_x_l, b_x_r)
    o_y_l, o_y_r = compare_dim(a_y_l, a_y_r, b_y_l, b_y_r)
    
    if (o_x_l != -1) and (o_y_l != -1):
       #print('overlap coord. x = ({}, {}), y = ({}, {})'.format(o_x_l, o_x_r, o_y_l, o_y_r))
       return o_x_l, o_x_r, o_y_l, o_y_r
    else:
       return 0, 0, 0, 0

# print initial 5 claims
print_claims(claims[0:3])

# main
print('total claims = {}'.format(claim_num)) # 1233 claims
area = []
for j in range(0, claim_num):
    for i in range(j+1,claim_num):
        #print('compare claims[0] against claims[{}]'.format(i))
        o_x_l, o_x_r, o_y_l, o_y_r = compare_claims(claims[j], claims[i])
        if o_x_l != 0:
            area.append([o_x_l, o_x_r, o_y_l, o_y_r])

#print('total_area = {}'.format(area))
print('max_x = {}, max_y = {}'.format(max_x, max_y))

# create 2-dim matrix (todo: write a numpy oriented alternative)
m=[]
for x in range(max_x + 1):
    m.append([]) # empty list
    for y in range(max_y + 1):
        m[x].append(0) # append indiv item

for k in area:
    for u in range(k[0], k[1]+1): # x dimension
        for v in range(k[2], k[3]+1):
             m[u][v] = 1
    
print('areas = {}'.format(len(area))) # 1758 areas

sum = 0
for x in range(max_x + 1):
    for y in range(max_y + 1):
        sum += m[x][y]

print('sum = {}'.format(sum)) # correct answer is: 101196 
```
