```python
filename = "aoc_2018_2.txt"

# part 1: 
two_code = 0
three_code = 0

file = open(filename, "r")
for line in file:
    print(line)
    # find pair of 2 
    
    match = {}
    # read each character
    for i in line:
        #print(i)
        if i in match:
           match[i] += 1
        else:
           match[i] = 1
    # count pairs of 2 and pairs of 3
    two_cnt = 0
    three_cnt = 0 
    for k,v in match.items():
        print('k={}, v={}'.format(k,v))
        if v%2 == 0: # test if v is a multiple of 2
            two_cnt += 1
        if v%3 == 0: # test if v is a multiple of 2
            three_cnt += 1
            
    print('two_cnt={}'.format(two_cnt))
    print('three_cnt={}'.format(three_cnt))
            
    if two_cnt > 0:
        two_code += 1
    if three_cnt > 0:
        three_code += 1
        
    print('two_code={}'.format(two_code))
    print('three_code={}'.format(three_code))
        
print(two_code * three_code) # part 1 answer: 6000   
```

```python
def cmp_codes(i, j):
    """
    Compare codes for equality, allowing for no more than 1 difference in the same position.
    
    Parameters
    ----------
        i : str
            first input string
        j : str
            second input string 
        
    Returns
    -------
        boolean 
            True - match, False - mis-match
    """
    mis_match = False
    for k in range(len(i)):
        if(i[k] == j[k]):
            pass
        else:
            # allow for one mis_match
            if mis_match == True:
                break
            mis_match = True
    if k == len(i) -1:
        return(True)
    else:
        return(False)
    
def common_char(i, j):
    """
    Identify the common characters for each character position in the string
    
    Parameters
    ----------
        i : str
            first input string
        j : str
            second input string
            
    Returns
    -------
        str 
            string with common characters
    """
    r = []
    for k in range(len(i)):
        if(i[k] == j[k]):
            r.append(i[k])
    return ''.join(r)

def tests():
    i = 'pbokrmjeiquhxlqnywsfgtucdv\n'
    j = 'pbokrmjeiquhxlqnywefgtycdv\n'
    assert cmp_codes(i,j) == False
    i = 'pbokrmjeiquhxlqnywsfgtucdv\n'
    j = 'pbokrmjeiquhxlqnywsfgtycdv\n'
    assert cmp_codes(i,j) == True
        
    assert common_char(i, j) == 'pbokrmjeiquhxlqnywsfgtcdv\n'
    
tests() # no output means tests have passed    
```
