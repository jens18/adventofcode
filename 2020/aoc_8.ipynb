```python
# AOC 2020: https://adventofcode.com/2020/day/8

asm_sample="""nop +0
acc +1
jmp +4
acc +3
jmp -3
acc -99
acc +1
jmp -4
acc +6
"""

def parse_asm(asm,prg):
    for l in asm.splitlines():
        instruct, arg = l.split(" ")
        print("instruct=",instruct,",arg=",arg) 
        prg.append([instruct, int(arg), 0]) # list format: instruction, arg, (execution-)count
        
def interpret(prg):
    idx = 0
    acc = 0

    while True:
        s = prg[idx]
        print("before: s=",s,"idx=",idx,"acc=",acc)
        if s[0] == 'nop':
            if s[2] == 0:
                idx=idx+1
                s[2]=1
            else:
                break
        elif s[0] == 'jmp':
            if s[2] == 0:
                idx=idx+s[1]
                s[2]=1
            else:
                break
        elif s[0] == 'acc':
            if s[2] == 0:
                idx=idx+1
                acc = acc+s[1]
                s[2]=1
            else:
                break
        print("after: s=",s,"idx=",idx,"acc=",acc)
                
    return(acc)       
            
prg = []
parse_asm(asm_sample,prg)
print(prg)
acc = interpret(prg)
print("acc=",acc)
```

```python
# part 1

file = open("aoc_8.txt", "r")
#read whole file to a string
asm = file.read()
file.close()

prg = []
parse_asm(asm,prg)
acc = interpret(prg)
print("acc=",acc)
```
