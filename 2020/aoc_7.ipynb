```python
# AOC 2020: https://adventofcode.com/2020/day/7

bag_rules_sample="""light red bags contain 1 bright white bag, 2 muted yellow bags.
dark orange bags contain 3 bright white bags, 4 muted yellow bags.
bright white bags contain 1 shiny gold bag.
muted yellow bags contain 2 shiny gold bags, 9 faded blue bags.
shiny gold bags contain 1 dark olive bag, 2 vibrant plum bags.
dark olive bags contain 3 faded blue bags, 4 dotted black bags.
vibrant plum bags contain 5 faded blue bags, 6 dotted black bags.
faded blue bags contain no other bags.
dotted black bags contain no other bags.
"""

# create 'container_bag_tag' hashtable of set with 'contained_bag_tag'
 
def parse_bag_rules(bags, bag_rules):
  for l in bag_rules.splitlines():
    container_bag_tag, content = l.split(' bags contain ')
    print("*** container_bag_tag=",container_bag_tag)
    contained_bags = content.split(', ')
    #print(contained_bags)
    bags_val = set()
    for b in contained_bags:
        if b != 'no other bags.':
            bag_elements = b.split(' ')
            #print("bag_elements=", bag_elements)
            contained_bag_tag=bag_elements[1] + " " + bag_elements[2]
            print("contained_bag_tag=",contained_bag_tag)
            bags_val.add(contained_bag_tag)
    bags[container_bag_tag] = bags_val
  
bags = {}
parse_bag_rules(bags, bag_rules_sample)
print(bags)   
```

```python
# recursion
def find_bag_rules(match, result):
    for i in bags.keys():
        #print(i)
        val = bags[i] 
        if match in val:
            result[i] = 1
            print("container bag=", i, " contains=", match)
            # continue search with container bag; recursion stops when no more matches are found
            find_bag_rules(i, result)
 
result = {}
find_bag_rules('shiny gold', result)
print(len(result))
```

```python
# part 1

file = open("aoc_7.txt", "r")
#read whole file to a string
bag_rules = file.read()
file.close()

bags = {}
parse_bag_rules(bags, bag_rules)
print(bags)

result = {}
find_bag_rules('shiny gold', result)
print(len(result))

```
